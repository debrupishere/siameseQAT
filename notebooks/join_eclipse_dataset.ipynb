```python
import pandas as pd
```

```python
path_eclipse_small = 'data/normalized/eclipse_small/eclipse_small.csv'
path_eclipse_small_pairs = 'data/normalized/eclipse_small/eclipse_small_pairs.csv'
path_eclipse = 'data/normalized/eclipse/eclipse.csv'
path_eclipse_pairs = 'data/normalized/eclipse/eclipse_pairs.csv'
```

```python
df_eclipse_small = pd.read_csv(path_eclipse_small, error_bad_lines=False)
df_eclipse_small_pairs = pd.read_csv(path_eclipse_small_pairs, error_bad_lines=False)
df_eclipse = pd.read_csv(path_eclipse)
df_eclipse_pairs = pd.read_csv(path_eclipse_pairs)
```

```python
print("Total eclipse small: {}".format(len(df_eclipse_small)))
print("Total eclipse: {}".format(len(df_eclipse)))
```

```python
print("Total eclipse small pairs: {}".format(len(df_eclipse_small_pairs)))
print("Total eclipse pairs: {}".format(len(df_eclipse_pairs)))
```

```python
n_master_small = len(df_eclipse_small[df_eclipse_small['dup_id'] == '[]'])
n_master = len(df_eclipse[df_eclipse['dup_id'] == '[]'])

print("Total master small: {}".format(n_master_small))
print("Total master: {}".format(n_master))
```

```python
n_dup_small = len(df_eclipse_small[df_eclipse_small['dup_id'] != '[]'])
n_dup = len(df_eclipse[df_eclipse['dup_id'] != '[]'])

print("Total dup small: {}".format(n_dup_small))
print("Total dup: {}".format(n_dup))
```
